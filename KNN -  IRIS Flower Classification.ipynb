```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/umangkejriwal1122/Youtube/master/Iris.csv")
```

```python
data.head()
```

```python
x = data.iloc[:,1:5].values
y = data.iloc[:,5].values
```

```python
x
```

# Splitting into Training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20)
```

```python
xtrain
```

```python
len(xtrain)
```

# Scaling the data (Decreasing range of data i.e., if range is (1-100), then make it (0-1)

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.fit_transform(xtest)
```

```python
xtrain
```

# Build the model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=5)   ### n_neighbors is number of surrounder items. 
#### In this, n_neighbors = sqrt(len(xtrain))/2 = sqrt(120)/2 = 5
```

```python
model.fit(xtrain,ytrain)
```

```python
y_pred = model.predict(xtest)
```

```python
y_pred
```

```python
ytest
```

# Finding confusion matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm = confusion_matrix(ytest,y_pred)
ac = accuracy_score(ytest,y_pred)
```
